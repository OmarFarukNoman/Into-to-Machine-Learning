```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# For HW2, we're using housing dataset as per instruction.
url = "https://raw.githubusercontent.com/HamedTabkhi/Intro-to-ML/main/Dataset/Housing.csv"
h_data = pd.read_csv(url)

# I've defined i/p and o/p for this problem with ALL specified features
features = ['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad', 'guestroom',
           'basement', 'hotwaterheating', 'airconditioning', 'parking', 'prefarea']

X_in = h_data[features].copy()
y_out = h_data['price'].values.reshape(-1, 1)

# There are some categorical values. I've changed them into numerical value
categorical_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
for col in categorical_cols:
    X_in[col] = X_in[col].map({'yes': 1, 'no': 0})

X_in = X_in.values

# I've splitted the dataset into training (80%) and validation (20%)
X_train, X_val, y_train, y_val = train_test_split(X_in, y_out, test_size=0.2, random_state=80)

# Gradient descent implementation with intercept
def gradient_descent(X, y, X_val, y_val, learning_rate, max_iterations):
    m, n = X.shape
    X_b = np.c_[np.ones((m, 1)), X]

    m_val = X_val.shape[0]
    X_val_b = np.c_[np.ones((m_val, 1)), X_val]

    theta = np.zeros((n + 1, 1))
    train_losses = []
    val_losses = []

    for iteration in range(max_iterations):
        gradients = (1/m) * X_b.T @ (X_b @ theta - y)
        theta -= learning_rate * gradients

        train_loss = (1/(2*m)) * np.sum((X_b @ theta - y)**2)
        val_loss = (1/(2*m_val)) * np.sum((X_val_b @ theta - y_val)**2)

        train_losses.append(train_loss)
        val_losses.append(val_loss)

    return theta, train_losses, val_losses

learning_rate = 0.01
max_iterations = 1000

theta, train_losses, val_losses = gradient_descent(X_train, y_train, X_val, y_val, learning_rate, max_iterations)

# I've plotted the figure
plt.figure(figsize=(6, 4))
plt.plot(train_losses, label='Training Loss', linewidth=2, color='red')
plt.plot(val_losses, label='Validation Loss', linewidth=2, linestyle='--', color='black')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title(f'Learning Rate: {learning_rate} (Input features:11)')
plt.legend()
plt.show()

# Results
print(f"Learning Rate: {learning_rate}")
print(f"Final Training Loss: {train_losses[-1]:.2f}")
print(f"Final Validation Loss: {val_losses[-1]:.2f}")
print(f"Theta values: {theta.flatten()}")

# Best parameters
print("\nBest Parameters:")
print(f"Theta_0: {theta[0][0]:.2f}")
for i, feature in enumerate(features):
    print(f"{feature} (theta_{i+1}): {theta[i+1][0]:.2f}")
```
